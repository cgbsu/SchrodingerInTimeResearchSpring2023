```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
from pathlib import Path
```

```python
from typing import List
```

```python
name : str = "top_down_stair_2d_test_figs"
```

```python
data = pd.read_csv(Path(name) / (name + ".csv"))
```

```python
count = 4
```

```python
postRegionProbabilities : List[np.array] = []
intraRegionProbabilities : List[np.array] = []
preRegionProbabilities : List[np.array] = []
for ii in range(count): 
     postRegionProbabilities.append(data[name + str(ii) + "::Probabilities::Post"])
     #intraRegionProbabilities.append(data[name + str(ii) + "::Probabilities::Intra"])
     preRegionProbabilities.append(data[name + str(ii) + "::Probabilities::Pre"])
```

```python
plt.figure()
legendLabels = []
ii = 0
for probabilities in postRegionProbabilities: 
    plt.plot(np.arange(len(postRegionProbabilities[0])), probabilities)
    legendLabels.append("Post Region" + str(ii))
    ii += 1
plt.title(name)
plt.legend(legendLabels)
```

```python
plt.figure()
legendLabels = []
ii = 0
for probabilities in preRegionProbabilities: 
    plt.plot(np.arange(len(postRegionProbabilities[0])), probabilities)
    legendLabels.append("Pre Region" + str(ii))
    ii += 1
plt.title(name)
plt.legend(legendLabels)
```
