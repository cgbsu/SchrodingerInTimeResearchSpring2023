```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
control : pd.DataFrame = pd.read_csv("Tunnlingx3StaticStairwell0.csv")
```

```python
variable : pd.DataFrame = pd.read_csv("StaticStairwell3xBarrier0.csv")
```

```python
controlPostBarrierProbabilities = np.array(control["Post-Barrier-Probabilities"])
```

```python
controlIntraBarrierProbabilities = np.array(control["Intra-Barrier-Probabilities"])
```

```python
controlPreBarrierProbabilities = np.array(control["Pre-Barrier-Probabilities"])
```

```python
variablePostBarrierProbabilities = np.array(variable["Post-Barrier-Probabilities"])
```

```python
variableIntraBarrierProbabilities = np.array(variable["Intra-Barrier-Probabilities"])
```

```python
variablePreBarrierProbabilities = np.array(variable["Pre-Barrier-Probabilities"])
```

```python
plt.figure()
plt.title("Probabilities Control and Variable for Intra and Post Barrier Regions")
plt.plot(controlPostBarrierProbabilities)
plt.plot(controlIntraBarrierProbabilities)
plt.plot(controlPostBarrierProbabilities + controlIntraBarrierProbabilities)
plt.plot(variablePostBarrierProbabilities)
plt.plot(variableIntraBarrierProbabilities)
plt.plot(variablePostBarrierProbabilities + variableIntraBarrierProbabilities)
#plt.plot(variablePostBarrierProbabilities / controlPostBarrierProbabilities)
#plt.plot((variablePostBarrierProbabilities + variableIntraBarrierProbabilities) \
        #/ (controlPostBarrierProbabilities + controlIntraBarrierProbabilities))
plt.xlabel("Time")
plt.ylabel("Probability/Probability Ratio")
plt.legend([
            "Control: Post-Barrier Probabilities", 
            "Control: Intra-Barrier Probabilities", 
            "Control: Intra- + Post- Barrier Probabilities", 
            "Variable: Post-Barrier Probabilities", 
            "Variable: Intra-Barrier Probabilities", 
            "Variable: Intra- + Post- Barrier Probabilities", 
            #"Varible Post-Barrier Probabilities/Control Post-Barrier Probabilities", 
            #"Variable Intra+Post-Barrier Probabilities/Control Intra+Post-Barrier Probabilities"
        ], 
        bbox_to_anchor = (1.1, 1.05)
    )
```

```python
plt.figure()
plt.title("Variable Post-Barrier Probabilities / Control Post-Barrier Probabilities")
plt.xlabel("Time")
plt.ylabel("Probability Ratio")
plt.plot(variablePostBarrierProbabilities / controlPostBarrierProbabilities)
```

```python
plt.figure()
plt.title("Variable Post-Barrier Probabilities / Control Post-Barrier Probabilities (First 100 Time Intervals)")
plt.xlabel("Time")
plt.ylabel("Probability Ratio")
plt.plot(variablePostBarrierProbabilities[:100] / controlPostBarrierProbabilities[:100])
```

```python
plt.figure()
plt.title("Variable Intra+Post-Barrier Probabilities/Control Intra+Post-Barrier Probabilities")
plt.xlabel("Time")
plt.ylabel("Probability Ratio")
plt.plot((variablePostBarrierProbabilities + variableIntraBarrierProbabilities) \
        / (controlPostBarrierProbabilities + controlIntraBarrierProbabilities))
```

```python
plt.figure()
plt.title("Variable Intra+Post-Barrier Probabilities/Control Intra+Post-Barrier Probabilities (First 100 Time Intervals)")
plt.xlabel("Time")
plt.ylabel("Probability Ratio")
plt.plot((variablePostBarrierProbabilities + variableIntraBarrierProbabilities)[:100] \
        / (controlPostBarrierProbabilities + controlIntraBarrierProbabilities)[:100])
```

```python
plt.figure()
plt.title("Variable Intra-Barrier Probabilities / Control Intra-Barrier Probabilities")
plt.xlabel("Time")
plt.ylabel("Probability Ratio")
plt.plot(variableIntraBarrierProbabilities / controlIntraBarrierProbabilities)
```

```python
plt.figure()
plt.xlabel("Time")
plt.ylabel("Probability Ratio")
plt.title("Variable Post-Barrier Probabilities / Control Intra-Barrier Probabilities")
plt.plot(variablePostBarrierProbabilities / controlIntraBarrierProbabilities)
plt.plot(controlIntraBarrierProbabilities)
plt.plot(variablePostBarrierProbabilities)
plt.legend(["Ratio", "Control", "Variable"])
```

```python
plt.figure()
plt.xlabel("Time")
plt.ylabel("Probability Ratio")
plt.title("Variable Intra-Barrier Probabilities / Control Post-Barrier Probabilities")
plt.plot(variableIntraBarrierProbabilities / controlPostBarrierProbabilities)
```

```python
plt.figure()
plt.xlabel("Time")
plt.ylabel("Probability Ratio")
#plt.plot((variablePostBarrierProbabilities / controlIntraBarrierProbabilities)[:200])
#plt.plot(np.log10(variableIntraBarrierProbabilities / controlPostBarrierProbabilities)[:200])
plt.plot(.5 * (variablePostBarrierProbabilities / controlPostBarrierProbabilities)[:200])
plt.plot((variablePreBarrierProbabilities / controlPreBarrierProbabilities)[:200])
plt.plot((variableIntraBarrierProbabilities / controlIntraBarrierProbabilities)[:200])
plt.legend([
        #"Variable Post-Barrier Probabilities / Control Intra-Barrier Probabilities", 
        #"Variable Intra-Barrier Probabilities / Control Post-Barrier Probabilities", 
        "Variable Post-Barrier Probabilities / Control Post-Barrier Probabilities", 
        "Variable Pre-Barrier Probabilities / Control Pre-Barrier Probabilities", 
        "Variable Intra-Barrier Probabilities / Control Intra-Barrier Probabilities", 
    ], 
    bbox_to_anchor = (1.1, 1.05)
)
```

```python
plt.figure()
plt.xlabel("Time")
plt.ylabel("Probability Ratio")
plt.title("Variable Post-Barrier Probabilities / Control Post-Barrier Probabilities")
#plt.plot(.1*((variablePostBarrierProbabilities / controlPostBarrierProbabilities) - 1)[:200])
plt.plot(variablePostBarrierProbabilities[0:30])
plt.plot(controlPostBarrierProbabilities[0:30])
plt.axhline(0)
plt.legend([
        #"Variable Post-Barrier Probabilities / Control Post-Barrier Probabilities", 
        "Variable Post-Barrier Probabilities", 
        "Control Post-Barrier Probabilities"
    ], 
    bbox_to_anchor = (1.1, 1.05)
)
        
```

```python
plt.figure()
plt.xlabel("Time")
plt.ylabel("Probability Ratio")
plt.title("Variable Post-Barrier Probabilities / Control Post-Barrier Probabilities")
#plt.plot(.1*((variablePostBarrierProbabilities / controlPostBarrierProbabilities) - 1)[:200])
plt.plot(variablePostBarrierProbabilities[0:60])
plt.plot(controlPostBarrierProbabilities[0:60])
plt.axhline(0)
plt.legend([
        #"Variable Post-Barrier Probabilities / Control Post-Barrier Probabilities", 
        "Variable Post-Barrier Probabilities", 
        "Control Post-Barrier Probabilities"
    ], 
    bbox_to_anchor = (1.1, 1.05)
)
        
```
